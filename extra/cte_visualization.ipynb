```python
import plotly.offline as py
from plotly.graph_objs import *
import pandas as pd
import math
import os
import numpy as np

py.init_notebook_mode()


my_cols=['speed','angle', 'cte', 'steer_value', 'throttle']

def read_file(file_name):
    with open(os.path.join('..', 'build', file_name)) as f:
        table_out = pd.read_table(f, sep='\t', header=None, names=my_cols, lineterminator='\n')
    return table_out

```

```python
import re

p = re.compile('cte_out_(.*)_(.*)_(.*).txt')

def plot_cte(file_name, start = 0, end = None):
        
    m = p.match(file_name)
    
    kP = m.group(1)
    kI = m.group(2)
    kD = m.group(3)
    
    title = 'CTE (kP: {}, kI: {}, kD: {})'.format(kP, kI, kD)
    
    table = read_file(file_name)
    
    values = table['cte'][start:end]
    mean = np.mean(values)
    
    t = list(range(0, len(values)))
    
    trace2 = Scatter(
        x=t,
        y=values,
        xaxis='x2',
        yaxis='y2',
        name = 'CTE',
        #mode = 'markers'
    )


    trace1= Scatter(
        x=[t[0], t[-1]],
        y=[mean, mean],
        xaxis='x2',
        yaxis='y2',
        name = 'Mean Value',
        #mode = 'markers'
    )

    data = [trace2, trace1]

    layout = Layout(
        title=title,
        xaxis2=dict(
            anchor='x2',
            title='t'
        ),
        yaxis2=dict(
            anchor='y2',
            title='CTE'
        )
    )
    
    fig = Figure(data=data, layout=layout)
    
    py.iplot(fig, filename= 'CTE')
```

```python
plot_cte('cte_out_0.260047_0_5.51264.txt')
```
